#Missing Value

```python
#missing value detectionin rows
cleaning.isnull().any(axis='columns')
```

```python
#missing value detectionin columns
cleaning.isnull().any(axis='rows')
```

```python
#drop the unidentified customer ID column
cleaning = cleaning.dropna(subset=['CustomerID'])
cleaning
```

#Duplicate Value

```python
#drop the duplicate(s)
cleaning = cleaning.drop_duplicates()
cleaning
```

#Drop the <1 Quantity and <0 unit price

```python
cleaning = cleaning.loc[(cleaning['Quantity']>0) & (cleaning['UnitPrice']>0)]
cleaning
```

#Change Data Type

```python
cleaning.info()
```

```python
#change the Customer ID type
cleaning['CustomerID'] = cleaning['CustomerID'].astype(object)
cleaning.info()
```

```python
cleaning['InvoiceDate'] = pd.to_datetime(cleaning['InvoiceDate'])
cleaning.info()
```

```python
cleaning['InvoiceDate.1'] = pd.to_datetime(cleaning['InvoiceDate.1'])
cleaning.info()
```

```python
cleaning
```

#Rename Column

```python
cleaning = cleaning.rename(columns={'InvoiceDate.1': 'InvoiceTimestamp'})
cleaning.info()
```
